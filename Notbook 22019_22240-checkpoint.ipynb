**Objectif**

L'optimisation du taux de conversion est probablement le travail le plus courant d'un Rate scientist, et à juste titre.
La révolution des données a beaucoup à voir avec le fait que nous sommes maintenant en mesure de collecter toutes sortes de données sur les personnes qui achètent quelque chose sur notre site ainsi que sur celles qui n'achètent pas. Cela nous donne une opportunité énorme de comprendre ce qui fonctionne bien (et éventuellement de le développer encore davantage) et ce qui ne fonctionne pas bien (et de le corriger).
L'objectif de ce défi est de construire un modèle qui prédit le taux de conversion et, sur la base du modèle, trouver des idées pour améliorer les revenus.
Ce défi est nettement plus facile que tous les autres de cette collection. Il n'y a pas de dates, pas de tables à joindre, aucune ingénierie de fonction requise, et le problème est vraiment simple.
Par conséquent, c'est un excellent point de départ pour se familiariser avec les défis de Rate science à domicile.
Vous ne devriez pas passer aux autres défis tant que vous n'avez pas parfaitement compris celui-ci.

**Entrainer un modèle multivarié en ajoutant plus de variables à la liste des caractéristiques**

```python
features_list =['total_pages_visited', 'age', 'country', 'source', 'new_user']

numerical_features = ['total_pages_visited', 'age']
categorical_features = ['country', 'source', 'new_user']

target_variable = 'converted'

X = Rate.loc[:, features_list]
Y = Rate.loc[:, target_variable]

print('Explanatory variables : ', X.columns)
print()
```

**Diviser l'ensemble de données en ensemble d'entraînement et ensemble de test**

```python
print("Division en ensembles d'entraînement et de test...")
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=0)
print("...Terminé.")
print()
```

## Choix des variables à utiliser dans le modèle et création des ensembles d'entraînement et de test

**D'après l'analyse exploratoire des données (EDA), nous savons que la caractéristique la plus utile est total_pages_visited. Créons d'abord un modèle de base en utilisant uniquement cette caractéristique : dans les cellules suivantes, nous effectuerons des prétraitements et entraînerons une régression logistique simple (univariée).**

<font color = "navy"> <span style = "font-family : 'Times New Roman' ; font-size: 14pt" > <b>  Précision de référence
 </b>  </span></font>

```python
1-sum(Rate['converted'])/len(Rate['converted'])
```

**Créer un pipeline pour les caractéristiques numériques**

```python
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# Créer un pipeline pour les caractéristiques catégorielles
categorical_transformer = Pipeline(steps=[
    ('encoder', OneHotEncoder(drop='first')) # La première colonne sera supprimée pour éviter de créer des corrélations entre les caractéristiques
])

# Utiliser ColumnTransformer pour créer un objet de prétraitement qui décrit tous les traitements à effectuer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
print("Prétraitement de X_train...")
print()
X_train = preprocessor.fit_transform(X_train)
print("...Terminé !")
print(X_train[0:5,:]) # X_train est maintenant un tableau numpy
print()

print("Prétraitement de X_test...")

X_test = preprocessor.transform(X_test)
print("...Terminé !")
print(X_test[0:5,:]) # X_test est maintenant un tableau numpy
```

```python
print("Conversion des DataFrames pandas en tableaux numpy...")
Y_train = Y_train.values
Y_test = Y_test.values
print("...Terminé !")
print(Y_train[0:5])
print(Y_test[0:2])
```

<font color = "navy"> <span style = "font-family : 'Times New Roman' ; font-size: 16pt" > <b> <b> ## Training 
</b>  </span></font>

```python
# Entraînement du modèle
print("Entraînement du modèle...")
classifieur = LogisticRegression()
classifieur.fit(X_train, Y_train)
print("...Terminé.")
```

<font color = "navy"> <span style = "font-family : 'Times New Roman' ; font-size: 16pt" > <b> Predictions  set...<b>
</b>  </span></font>

```python
# Prédictions sur l'ensemble d'entraînement
print("Prédictions sur l'ensemble d'entraînement...")
Y_train_pred = classifieur.predict(X_train)
print("...Terminé.")
print(Y_train_pred)
print()
```

```python
# Prédictions sur l'ensemble de test
print("Prédictions sur l'ensemble de test...")
Y_test_pred = classifieur.predict(X_test)
print("...Terminé.")
print(Y_test_pred)
print()
```

```python
# AT
print("Score f1 sur l'ensemble d'entraînement : ", f1_score(Y_train, Y_train_pred))
print("Score f1 sur l'ensemble de test : ", f1_score(Y_test, Y_test_pred))
```

<font color = "navy"> <span style = "font-family : 'Times New Roman' ; font-size: 16pt" > <b> Visualiser les matrices de confusion<b>
</b>  </span></font>

```python
print("Matrice de confusion sur l'ensemble d'entraînement : ")
print(confusion_matrix(Y_train, Y_train_pred))
print()
print("Matrice de confusion sur l'ensemble de test : ")
print(confusion_matrix(Y_test, Y_test_pred))
print()
```

```python
_ , ax = plt.subplots() # Obtenir un sous-graphique de matplotlib
ax.set(title="Matrice de confusion sur l'ensemble d'entraînement") # Définir un titre que nous ajouterons à ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(classifieur, X_train, Y_train, ax=ax)
# ConfusionMatrixDisplay de sklearn
plt.savefig('Matrice de confusion entraînement.jpg')
plt.show()

_ , ax = plt.subplots() # Obtenir un sous-graphique de matplotlib
ax.set(title="Matrice de confusion sur l'ensemble de test") # Définir un titre que nous ajouterons à ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(classifieur, X_test, Y_test, ax=ax) # ConfusionMatrixDisplay de sklearn
plt.savefig('Matrice de confusion test.jpg')
plt.show()
```
